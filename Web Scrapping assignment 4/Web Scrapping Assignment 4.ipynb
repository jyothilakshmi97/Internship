```python
#1Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:A) Rank B) Name C) Artist D) Upload date E) Views

driver=webdriver.Chrome()
time.sleep(2)
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)

rank=[]
try:
    position=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]//tr//td[1]')
    for i in position:
        rank.append(i.text)
except NoSuchElementException:
    rank.append('-')
except StaleElementReferenceException:
    rank.append('-')
    
time.sleep(2)
name=[]
try:
    names=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]//tr//td[2]')
    for i in names:
        name.append(i.text)
except NoSuchElementException:
    name.append('No details available')
except StaleElementReferenceException:
    name.append('No details available')
    
artist=[]
try:
    ar=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]//tr//td[3]')
    for i in ar:
        artist.append(i.text)
except NoSuchElementException:
    artist.append('-')
except StaleElementReferenceException:
    artist.append('-')

upload=[]
try:
    date=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]//tr//td[5]')
    for i in date:
        upload.append(i.text)
except NoSuchElementException:
    upload.append('-')
except StaleElementReferenceException:
    upload.append('-')
    
views=[]
try:
    view=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]//tr//td[4]')
    for i in view:
        views.append(i.text)
except NoSuchElementException:
    views.append('-')
except StaleElementReferenceException:
    views.append('-')
    
df=pd.DataFrame({"Rank":rank,"Name":name,"artist":artist,"Uploaded Date":upload,"Views(in billion)":views})
df
```

```python
name=[]
try:
    for i in range(0,3):
        names=driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]')
    for i in names:
        name.append(i.text)
except NoSuchElementException:
    name.append('No details available')
   
    
year=[]
try:
    info=driver.find_elements(By.XPATH,'//table[@class="col-span-full my-2 table sm:col-start-2"]//tbody//tr//td[3]')
    for i in info:
         year.append(i.text)
except NoSuchElementException:
    year.append("NA")


no_of_attributes=[]
try:
    info=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]//div[4]')
    for i in info:
         no_of_attributes.append(i.text)
except NoSuchElementException:
    no_of_attributes.append("NA")


no_of_instances=[]
try:
    info=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]//div[3]//span')
    for i in info:
         no_of_instances.append(i.text)
except NoSuchElementException:
    no_of_instances.append("NA")


attr_type=[]
try:
    info=driver.find_elements(By.XPATH,'//table[@class="col-span-full my-2 table sm:col-start-2"]//tbody//tr//td[2]')
    for i in info:
         attr_type.append(i.text)
except NoSuchElementException:
    attr_type.append("NA")


default_task=[]
try:
    info=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]//div[1]//span')
    for i in info:
        default_task.append(i.text)
except NoSuchElementException:
    default_task.append("NA")


data_type=[]
try:
    info=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]//div[2]//span')
    for i in info:
         data_type.append(i.text)
except NoSuchElementException:
    data_type.append("NA")


df=pd.DataFrame({'Dataset name':name,'Data type':
                 data_type,'Task':default_task,'Attributetype':attr_type,'No of instance':no_of_instances,
                 'No of attribute':no_of_attributes,'Year':year })
df
```
